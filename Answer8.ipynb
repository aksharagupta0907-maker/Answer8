```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu, sigma = 50, 10  # Mean and standard deviation
data = np.random.normal(mu, sigma, 1000)

# 2. Calculate the 95% Confidence Interval for the mean
sample_mean = np.mean(data)
sem = stats.sem(data)  # Standard Error of the Mean
confidence_level = 0.95
degrees_freedom = len(data) - 1

# Calculate interval using the t-distribution
ci_lower, ci_upper = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sem)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# 3. Plot the data using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulated Data')

# Add the theoretical normal distribution curve
x = np.linspace(min(data), max(data), 100)
p = stats.norm.pdf(x, mu, sigma)
plt.plot(x, p, 'r', linewidth=2, label='Normal Distribution Curve')

# Formatting the plot
plt.axvline(sample_mean, color='green', linestyle='dashed', linewidth=2, label=f'Mean: {sample_mean:.2f}')
plt.title('Histogram of Simulated Normal Distribution Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
